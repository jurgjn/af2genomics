### Yeast missense predictions
Pathogenicity scores using the method from [Brandes2023](https://doi.org/10.1038/s41588-023-01465-0):
- `esm_pathogenicity_score` ESM1b LLR score; more negative means more pathogenic (see [Fig2a](https://www.nature.com/articles/s41588-023-01465-0/figures/2) from above)
- `esm_pathogenicity_label` benign/pathogenic labels based on the ESM1b LLR score, thresholded at -7.5 as in the paper

FoldX stability:
- `foldx_total_energy` ddG score prediting effect on stability
- `foldx_label` thresholded ddG score - `destabilizing` if ddG > 2; `no_call` means low-quality structure
- Other `foldx_...` columns are raw output from foldx pssm describing e.g. individual terms of the force field
